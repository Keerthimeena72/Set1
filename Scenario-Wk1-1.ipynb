```python
# A text-processing tool reverses a given word or sentence for formatting purposes.
# Write logic to take a word or sentence as input and produce its reversed version.
word=input("Enter a word : ")
rev_word=""
for i in range(len(word)-1,-1,-1):
    rev_word=rev_word + word[i]
print("Word          :",word)
print("Reversed word :",rev_word)
           
```

```python
#A grading system determines whether a student has passed or failed based on their score.
#Write logic to check if a student has passed a subject by scoring at least 40 marks.
eng_Mark=int(input("Enter your marks in English :"))
tam_Mark=int(input("Enter your marks in Tamil : "))
mat_Mark=int(input("Enter your marks in Maths :"))   
soc_Mark=int(input("Enter your marks in Social Studies :"))
sci_Mark=int(input("Enter your marks in Science :"))
if eng_Mark >= 40:
   print("English        : Passed") 
else:
   print("English        : Failed") 

if tam_Mark >= 40:
   print("Tamil          : Passed") 
else:
   print("Tamil          : Failed") 
   
if mat_Mark >= 40:
   print("Maths          : Passed") 
else:
   print("Maths          : Failed") 

if soc_Mark >= 40:
   print("Social Studies : Passed") 
else:
   print("Social Studies : Failed") 

if sci_Mark >= 40:
   print("Science        : Passed") 
else:
   print("Science        : Failed") 
```

```python
#A banking system processes withdrawal requests and ensures the user has enough balance.
#Write logic to check if a user has enough balance before allowing a withdrawal and update the remaining balance accordingly.
balance_Amt = 10000
withdrawal_Amt=int(input("Enter the withdrawal amount"))
if (withdrawal_Amt > balance_Amt):
    print("Sorry.....Not enough balance")
else : 
    balance_Amt = balance_Amt - withdrawal_Amt
    print("Successful withdrawal of ", withdrawal_Amt)
```

```python
#A calendar system verifies whether a given year is a leap year based on standard leap year rules.
# Write logic to determine whether a given year is a leap year.
input_Year=int(input("Enter the year :"))
if((input_Year %4 == 0) and (input_Year % 100 != 0) or (input_Year %400 == 0 )):
   print(input_Year, "is a leap year")
else :
   print(input_Year, "is not a leap year")
```
